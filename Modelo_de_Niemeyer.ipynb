```python
import numpy as np

def calculate_defect_params(defect_type, E_peak, cavity_size=1e-3, delta=1.0):
    """
    Calcula os campos de incepção e extinção para diferentes tipos de defeitos dielétricos.
    Baseado em características típicas descritas por Niemeyer e literatura de PD.
    """
    if defect_type == 'internal_void':
        # Cavidade interna (padrão simétrico)
        E_i = 3.0e6 * delta * (1 + 0.1 * np.log10(cavity_size / 1e-3))
        E_e = 0.1e6 * delta
        asymmetry = 1.0 # Simétrico
    elif defect_type == 'surface_discharge':
        # Descarga superficial (padrão assimétrico, geralmente maior no semiciclo negativo)
        E_i = 2.0e6 * delta
        E_e = 0.05e6 * delta
        asymmetry = 1.2 # Leve assimetria entre semiciclos
    elif defect_type == 'corona_point':
        # Descarga de corona / ponta (altamente assimétrica, ocorre perto do pico)
        E_i = 1.5e6 * delta
        E_e = 0.8e6 * delta
        asymmetry = 2.0 # Forte assimetria (mais descargas no semiciclo negativo)
    else:
        # Padrão genérico
        E_i = 2.5e6 * delta
        E_e = 0.1e6 * delta
        asymmetry = 1.0

    return E_i, E_e, asymmetry

def calculate_niemeyer_fields(E_cavity, pressure=101.3e3, temperature=293.15, cavity_size=1e-3, gas_type='air'):
    """
    Mantido para compatibilidade com versões anteriores.
    """
    delta = pressure / (101.3e3 * (temperature / 293.15))
    E_i = 3.0e6 * delta * (1 + 0.1 * np.log10(cavity_size / 1e-3))
    E_e = 0.1e6 * delta
    return E_i, E_e
```
